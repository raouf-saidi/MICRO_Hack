```python
import numpy as np
import pandas as pd

# nombre total d'instances
n = 10000

# génération de l'attribut tranche_age
tranche_age = np.random.choice([1, 2, 3], size=n, p=[0.45, 0.40, 0.15])
tranche_age = pd.Series(tranche_age, name='tranche_age')

# génération de l'attribut type_emp
type_emp = np.random.choice([1, 2, 3, 4], size=n, p=[0.35, 0.3, 0.25, 0.1])
type_emp = pd.Series(type_emp, name='type_emp')

# génération de l'attribut salaire_emp
salaire_emp = np.random.normal(loc=62500, scale=15000, size=n)
salaire_emp = np.clip(salaire_emp, a_min=25000, a_max=100000)
salaire_emp = pd.Series(salaire_emp, name='salaire_emp')

# génération de l'attribut nb_maladie
nb_maladie = np.random.exponential(scale=1/0.75, size=n)
nb_maladie = np.clip(nb_maladie.round(), a_min=0, a_max=2)
# la valeur 0 constitue 4/5 du nombre d'instances
nb_maladie_counts = pd.Series(nb_maladie).value_counts(normalize=True)
if nb_maladie_counts.get(0, 0) < 0.8:
    nb_maladie += np.random.choice([0, 1, 2], size=n, p=[0.8, 0.1, 0.1]) - nb_maladie
nb_maladie = pd.Series(nb_maladie, name='nb_maladie')

# génération de l'attribut durée_trv
durée_trv = np.random.choice([1, 0], size=n, p=[0.9, 0.1])
durée_trv = pd.Series(durée_trv, name='durée_trv')

# génération de l'attribut lieu_emp
lieu_emp = np.random.exponential(scale=1/0.4, size=n)
# l'attribut tranche_age est inversement corrélé avec lieu_emp
lieu_emp = np.where(tranche_age == 1, np.clip(lieu_emp, a_min=0, a_max=1), np.clip(lieu_emp, a_min=0, a_max=4))
# l'attribut type_emp est inversement corrélé avec lieu_emp
lieu_emp = np.where(type_emp == 4, np.clip(lieu_emp, a_min=0, a_max=1), np.clip(lieu_emp, a_min=0, a_max=4))
lieu_emp = pd.Series(lieu_emp, name='lieu_emp')
lieu_emp = np.clip(lieu_emp.round(), a_min=0, a_max=2)
# création de la dataset
dataset = pd.concat([tranche_age, type_emp, salaire_emp, nb_maladie, durée_trv, lieu_emp], axis=1)
```

```python
import numpy as np
import pandas as pd

# Définition des proportions pour chaque catégorie
proportions_age = [0.45, 0.40, 0.15]
proportions_type_emp = [0.40, 0.30, 0.20, 0.10]

# Nombre total d'instances
N = 1000

# Génération des tranches d'âge
tranche_age = np.random.choice([1, 2, 3], size=N, p=proportions_age)

# Génération des types d'emploi
type_emp = np.random.choice([1, 2, 3, 4], size=N, p=proportions_type_emp)

# Calcul des salaires en fonction des tranches d'âge et des types d'emploi
salaire_emp = np.zeros(N)
for i in range(N):
    if tranche_age[i] == 1:
        salaire_emp[i] = np.random.normal(loc=50000, scale=10000)
    elif tranche_age[i] == 2:
        salaire_emp[i] = np.random.normal(loc=70000, scale=15000)
    else:
        salaire_emp[i] = np.random.normal(loc=90000, scale=20000)
    if type_emp[i] == 2:
        salaire_emp[i] -= 5000
    elif type_emp[i] == 3:
        salaire_emp[i] += 10000
    elif type_emp[i] == 4:
        salaire_emp[i] += 20000

# Génération du nombre de maladies
nb_maladie = np.random.exponential(scale=0.5, size=N)
nb_maladie[nb_maladie > 2] = 2  # On ne garde que les valeurs entre 0 et 2
nb_maladie[0:int(0.8*N)] = 0  # On ajoute la contrainte sur le nombre de zéros
nb_maladie = np.clip(nb_maladie.round(), a_min=0, a_max=2)

# Génération de la durée de travail
duree_trv = np.random.choice([0, 1], size=N, p=[0.1, 0.9])

# Génération du lieu d'emploi
lieu_emp = np.random.exponential(scale=2, size=N)
# Génération du lieu d'emploi
lieu_emp = np.clip(lieu_emp.round(), a_min=0, a_max=2)
if len(np.where(lieu_emp > 10)[0]) > 0:
    lieu_emp[lieu_emp > 10] = 10  # On tronque la distribution pour éviter des valeurs extrêmes

# Création du dataframe
data = pd.DataFrame({
    'tranche_age': tranche_age,
    'type_emp': type_emp,
    'salaire_emp': salaire_emp,
    'nb_maladie': nb_maladie,
    'duree_trv': duree_trv,
    'lieu_emp': lieu_emp
})

# Calcul des corrélations entre les variables
corr = data.corr()
print(corr)
```
